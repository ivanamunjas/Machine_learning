```python
import pandas as pd 
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from matplotlib import pyplot as plt
%matplotlib inline

data = pd.read_csv("data.csv", header = None)
data.head()

X = data.loc[:, 0:4]
y = data.loc[:, 5]

X = np.array(X)
y = np.array(y)

X = (X - X.mean())/X.std()

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
X_train = np.c_[np.ones(len(X_train),dtype='int64'),X_train]
X_test = np.c_[np.ones(len(X_test),dtype='int64'),X_test]
```

```python
def my_train_test_split(X, Y, test_size=0.2, random_state=42):
    rnd = np.random.RandomState(random_state)
    X = np.array(X)
    Y = np.array(Y)
    ind = np.arange(len(X))
    rnd.shuffle(ind) 
    num = round((1-test_size)*len(ind))
    X_train = np.array([X[::][x] for x in ind[0:num]])
    X_test = np.array([X[::][x] for x in ind[num::]])
    Y_train = np.array([Y[::][y] for y in ind[0:num]])
    Y_test = np.array([Y[::][y] for y in ind[num::]])
    
    return X_train, X_test, Y_train, Y_test
```

```python
X = data.loc[:, 0:4]
y = data.loc[:, 5]

X = np.array(X)
y = np.array(y)

X = (X - X.mean())/X.std()

X_train, X_test, y_train, y_test = my_train_test_split(X, y, test_size=0.33, random_state=42)
X_train = np.c_[np.ones(len(X_train),dtype='int64'),X_train]
X_test = np.c_[np.ones(len(X_test),dtype='int64'),X_test]
```

## Linearna regresija

```python
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
y_pred = lin_reg.predict(X_test)

mean_squared_error(y_test, y_pred)
```

Moja linearna regresija ima slicnu strukturu kao linearna regresija iz sklearn-a. Ima fit metodu kojoj se prosledjuju vektor prediktora i vektor predikcija. Ovoj metodi se takodje prosledjuju dodatni parametri kao sto su _method_ koji odredjuje na koji nacin ce se proracunavati parametri $\theta$ i ima moguce vrednosti _stohastic_ i _batch_, na osnovu kojih se primenjuju __stochastic gradient descent__ ili __batch gradient descent__.

Ova klasa takodje sadrzi metodu __predict__ koja ima zadatak da na osnovu ulaznog testnog vektora prediktora proracuna predikciju y. 

Za poredjenje sa ugradjenim funkcijama implementirana je funkcija _my mse_ koja racuna __Mean squared error__. 

```python
class My_LinearRegression():
    def my_fit(self, X, y, iterations = 5000, alpha = 0.01, t0 = 5, t1 = 50, method = 'batch'):
        self.t0 = t0
        self.t1 = t1
        if method == 'batch':
            self.batch_gradient_descent(X, y, alpha, iterations)
        elif method == 'stohastic':
            self.stochastic_gradient_descent(X, y, iterations)
            
    def pred(self, X):
        return X.dot(self.theta)
    
    def batch_gradient_descent(self, X, Y, alpha, iterations):
        theta = np.zeros(X_train.shape[1])
        m = len(Y)
        for iteration in range(iterations):
            theta = theta - alpha * X.T.dot(X.dot(theta) - Y) / m

        self.theta = theta 
    
    def learning_schedule(self, t):
        return self.t0/(t + self.t1)

    def stochastic_gradient_descent(self, X, y, iterations):
        theta = np.random.randn(X_train.shape[1])
        m = X.shape[0]
        for iter_ in range(iterations):
            for i in range(X.shape[0]):
                random_index = np.random.randint(m)
                xi = X[random_index:random_index+1]
                yi = y[random_index:random_index+1]
                gradients = 2*xi.T.dot(xi.dot(theta) - yi)
                eta = self.learning_schedule(iter_*m + i)
                theta = theta - eta*gradients
                
        self.theta = theta
        
def my_mse(y_pred, y_test):
    return np.sum((y_pred - y_test)**2)/len(y_pred)
```

## Batch gradient descent

```python
my_lr = My_LinearRegression()
my_lr.my_fit(X_train, y_train)

y_pred = my_lr.pred(X_test)
print(my_mse(y_pred, y_test))
```

```python
my_lr = My_LinearRegression()
my_lr.my_fit(X_train, y_train, method = 'stohastic', iterations = 50)

y_pred = my_lr.pred(X_test)
print(my_mse(y_pred, y_test))
```

```python
# moja

class My_Ridge():
    def my_fit(self, X, y, alpha = 1):
        n = X.shape[0]
        m = X.shape[1]
        I = np.identity(m)
        self.w = np.dot(np.dot(np.linalg.inv(np.dot(X.T, X) + alpha * I), X.T), y)
    
    def pred(self, X):
        return X.dot(self.w)
```

```python
my_ridge = My_Ridge()
my_ridge.my_fit(X_train, y_train, 0.01)
y_pred = my_ridge.pred(X_test)
print(my_mse(y_pred, y_test))
```

```python
def pred(X, theta):
        return X.dot(theta)
    
def my_Lasso(X, y, alpha, iterations):
    theta = np.zeros(X.shape[1])
    m = y.shape[0]
    for iteration in range(iterations):
        theta = theta - alpha * (X.T.dot(X.dot(theta) - y) / m + np.sign(theta))

    return theta 
```

# Lokalno ponderisana regresija 

```python
def pred(X, theta):
    return X.dot(theta)

def my_mse(y_pred, y_test):
    return np.sum((y_pred - y_test)**2)/len(y_pred)

def localy_weighted_regression(X, Y, alpha, tau, iterations, xp):
    m = len(Y)
    W = np.zeros((m,1))
    theta = np.zeros((X.shape[1], 1))
    
    for i in range(m):
        W[i][0] = np.exp(-(np.linalg.norm(X[i] - xp)/tau)**2/2)   
        
    for iteration in range(iterations):
        theta = theta - alpha * X.T.dot(W*(X.dot(theta) - Y).reshape(m,1)) / m

    return theta
```

```python
alpha = 0.1
iter_ = 10000
tau = 1

y_ = y.reshape(len(y),1)

X_ = np.c_[np.ones(len(X),dtype='int64'),X]

y_pred = []
for i in range(X_.shape[0]):
    X_new = np.concatenate((X_[:i], X_[i+1:]))
    y_new = np.concatenate((y_[:i], y_[i+1:]))
    new_theta = localy_weighted_regression(X_new, y_new, alpha, tau, iter_, X_[i])

    y_pred.append(X_[i].dot(new_theta))


print(my_mse(np.array(y_pred), y_))
```

```python
for i in range(1, X_.shape[1]):
    plt.figure()
    plt.scatter(X_[:,i], y_, color = 'blue', alpha = 0.5)
    plt.plot(X_[:,i], X_[:,i]*new_theta[i] + new_theta[0], color='red')
    plt.legend(['Predikcija', 'Trening set', 'Test set'])
```

## Polinomijalna regresija

```python
# ugradjena 

poly_degree = [2, 3, 4, 5, 6]

from sklearn.preprocessing import PolynomialFeatures
import numpy as np

lin_reg = LinearRegression()

for p in poly_degree:
    poly_features = PolynomialFeatures(degree=p, include_bias=False)
    X_poly = poly_features.fit_transform(X_train)
    lin_reg.fit(X_poly, y_train)
    y_pred = lin_reg.predict(poly_features.fit_transform(X_test))
    print('Polinomial degree {}: '.format(p), mean_squared_error(y_test, y_pred))
```

```python
# moja

def poly_comb(features, deg):
    all_combs = []
    all_combs.append(list(features))
    chunks = np.arange(len(features))
    
    for i in range(0, deg-1):
        deg_combs = []
        for j in range(len(features)):
            next_chunk = len(deg_combs)
            for l in range(chunks[j], len(all_combs[i])):
                deg_combs.append(features[j]*all_combs[i][l])
            chunks[j] = next_chunk
        all_combs.append(deg_combs)
    
    all_combs = [j for i in all_combs for j in i]
    return all_combs
```

```python
X_poly = []
p = 2
for i in range(X.shape[0]):
    X_poly.append(poly_comb(X[i], p))

X_train, X_test, y_train, y_test = train_test_split(np.array(X_poly), y, test_size=0.33, random_state=42)
X_train = np.c_[np.ones(len(X_train),dtype='int64'),X_train]
X_test = np.c_[np.ones(len(X_test),dtype='int64'),X_test]

my_lr = My_LinearRegression()
my_lr.my_fit(X_train, y_train)
y_pred = my_lr.pred(X_test)
theta = my_lr.theta
print('Polinomial degree {}: '.format(p), my_mse(y_pred, y_test))
```
